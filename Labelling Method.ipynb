```python
#post tag kamus
kamus_pos = {'J':wordnet.ADJ, 'V':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}
#adj sebagai kata sifat (adjective)
#verb sebagai kata kerja 
#adv sebagai kata keterangan (adverb)
```

```python
from nltk.corpus import stopwords
import pycrfsuite

def token_stop_pos(text): 
    tags = pos_tag(word_tokenize(text))
    newlist =[]
    for word, tag in tags:
        if word.lower() not in set(stopwords.words('indonesian')):
            newlist.append(tuple([word, kamus_pos.get(tag[0])]))
            return newlist
```

```python
data['hasil pos tag']= df['opini'].apply(token_stop_pos)
data
```

# Lemmatization
Normalisasi kata 
teknik NLP untuk mengembalikan kata kepada kata dasarnya yang disesuaikan dengan kamus bahasa indonesia

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

#bikin function
def lemmatize(pos_data):
    lemma_r =" "
    for word, pos in pos_data:
        if not pos:
            lemma = word
            lemma_r = lemma_r + " " + lemma
        else :
            lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
            lemma_r =lemma_r + " " + lemma
    return lemma_r

data['Lemma'] = data['hasil pos tag'].apply(lemmatize)
```

```python
data.to_excel("hasil_label.xlsx")
```

# Text Blob

```python
from textblob import TextBlob

def getSubjectiv(review):
    return TextBlob(review).sentiment.subjectivity
def getpolarity(review):
    return TextBlob(review).sentiment.polarity
def analyze(score):
    if score > 0:
        return 'negatif'
    else :
        return 'positif'
    
```

```python
data['hasil_subjectivity'] = data['Lemma'].apply(getSubjectiv)
data['hasil_polaritas'] = data['Lemma'].apply(getpolarity)
data['hasil_analisis'] = data['hasil_polaritas'].apply(analyze)
```

# SentiWordnet

```python
def sentiwordnetAnalysis(pos_data):
    sentiment = 0
    tokens_count = 0
    for word, pos in pos_data:
        if not pos :
            continue
            lemma = wordnet_lemmatizer.lematize(word, pos=pos)
            if not lemma :
                continue
                
                synsets = wordnet.sysnset(lemma, pos=pos)
                if not synsets:
                    continue
                    
                    synsets = synsets[0]
                    swn_synsets = swn.senti_synset(synset.name())
                    sentiment += swn_synset.pos_score() - swn.synset.neg_score()
                    tokens_count += 1
                    
                    if not tokens_count:
                        return 0
                    
                    if sentiment > 0:
                        return 'Positif'
                    else :
                        return 'Negatif'
```

```python
data['sw'] = data['hasil pos tag'].apply(sentiwordnetAnalysis)
```
